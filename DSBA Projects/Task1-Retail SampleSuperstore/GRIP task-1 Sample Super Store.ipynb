### Findout Sum, mean, min, max, count, median, standard deviation, variance, of each state with respect to Sales

```python
state_sales = s_data.groupby('State').Sales.agg(['sum','mean','min','max','count','median','std','var']).sort_values(by='mean',ascending=False).reset_index()
state_sales
```

```python
plt.figure(figsize=(20,17))

sns.barplot(x='State',y='mean',data=state_sales)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xticks(rotation=90)
plt.xlabel('State',fontsize=30)
plt.ylabel('Mean Sales',fontsize=30)
plt.title('State wise Mean Sales',fontsize=40)
plt.show()
```

#### Observation:
    - Wyoming state have highest amount of sales from other states in US.
    - South Dakota have lowest amount of sales as compare to other states.

### Findout Sum, mean, min, max, count, median, standard deviation, variance, of each state with respect to Profit

```python
state_profit = s_data.groupby('State').Profit.agg(['sum','mean','min','max','count','median','std','var']).sort_values(by='mean',ascending=False).reset_index()
state_profit
```

```python
plt.figure(figsize=(20,17))

sns.barplot(x='State',y='mean',data=state_profit)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xticks(rotation=90)
plt.xlabel('State',fontsize=30)
plt.ylabel('Mean Profit',fontsize=30)
plt.title('State wise Mean Profit',fontsize=40)
plt.show()
```

#### Observation:
    - If we are checking for profit then 'Vermont' is having highest in profit making as compare to others.
    - On the other hand there are few states which has incurred loss. i.e. Ohio, Colorado, North Carolina, Texas,Arizona, Florida etc..
